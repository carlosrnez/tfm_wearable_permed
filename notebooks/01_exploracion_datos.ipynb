```python
import numpy as np

def explorar_npz_avanzado(ruta_archivo):
    print(f"--- Analisis profundo: {ruta_archivo} ---")
    datos = np.load(ruta_archivo, allow_pickle=True)
    
    arr_0 = datos['arr_0']
    arr_1 = datos['arr_1']
    arr_2 = datos['arr_2']
    
    print("\n1. VISTAZO A LOS DATOS (arr_0):")
    print(f"Primeros 5 valores de la fila 1: {arr_0[0][:5]}")
    
    print("\n2. VALORES ÚNICOS EN arr_1 (¿Actividades?):")
    # np.unique nos dice qué categorías hay y cuántas veces se repiten
    valores_1, cuentas_1 = np.unique(arr_1, return_counts=True)
    for v, c in zip(valores_1, cuentas_1):
        print(f"   - '{v}': {c} muestras")
        
    print("\n3. VALORES ÚNICOS EN arr_2 (¿Sujetos o dispositivos?):")
    valores_2, cuentas_2 = np.unique(arr_2, return_counts=True)
    # Mostramos solo los 10 primeros por si hay muchísimos
    for v, c in zip(valores_2[:10], cuentas_2[:10]): 
        print(f"   - '{v}': {c} muestras")
    if len(valores_2) > 10:
        print(f"   ... (y {len(valores_2) - 10} valores más)")

# Pon tu ruta exacta
ruta_de_tu_archivo = "C:/Users/CARLOS/Desktop/dev/tfm_wearablepermed/data/processed/sensor_M/data_feature_all_M.npz" 
explorar_npz_avanzado(ruta_de_tu_archivo)
```
